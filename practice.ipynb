# Import module

```python
import math
print(math.tau) # tau is the ratio of the circumference of a circle to its radius, which is 2*pi
print(math.pi) # pi is the ratio of the circumference of a circle to its diameter
print(math.cos(math.pi/4)) # cosine of 45 degrees is sqrt(2)/2
print(math.sin(math.pi/4)) # sine of 45 degrees is sqrt(2)/2
print(math.tan(math.pi/4)) # tangent of 45 degrees is 1
print(math.log(100,10)) # logarithm of 100 to the base 10
print(math.sqrt(2)) # square root
print(math.ceil(3.2)) # rounds up to the nearest integer
print(math.floor(3.8)) # rounds down to the nearest integer
print(math.factorial(5)) # 5! = 5*4*3*2*1
print(math.gcd(48,18)) # greatest common divisor
print(math.lcm(48,18)) # least common multiple
print(math.isqrt(10)) # integer square root
print(math.exp(1)) # euler's number
print(math.log2(8)) # logarithm of 8 to the base 2
print(math.log10(1000)) # logarithm of 1000 to the base
print(math.degrees(math.pi/4)) # convert radians to degrees
print(math.radians(45)) # convert degrees to radians
```

```python
my_list = [1, 2, 3, 4, 5]
i=0
while i<len(my_list):
    print(my_list[i])
    i+=1
```

# Matplotlib

```python
import numpy as np

from matplotlib import pyplot as plt

x_values = np.linspace(-3, 3, 100)

plt.figure()
plt.plot(x_values, np.sin(x_values), label="Sinusoid")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Matplotlib example")
plt.legend(loc="upper left")
plt.show()
```

     0, & \text{otherwise }.

```python
def manhattan_distance(point1, point2):
    return np.linalg.norm(np.array(point1) - np.array(point2), ord=1)
pointA = [1, 2]
pointB = [4, 6]
print(manhattan_distance(pointA, pointB)) # calculates the Manhattan distance between pointA and pointB
```

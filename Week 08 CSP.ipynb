```python
# Define a new cgraph
cgraph = {'WA': ['SA', 'NT'], 
          'Q': ['SA', 'NT', 'NSW'], 
          'T': [],
          'V': ['SA', 'NSW'], 
          'SA': ['WA', 'NT', 'Q', 'NSW','V'], 
          'NT': ['Q', 'WA', 'SA'], 
          'NSW': ['Q', 'V', 'SA']}
```

```python
import random
def get_color(node, rv, cg):
    # CODE
    # check rv[node] is empty or not when no color availabe
    # return 'no color'
    color = random.choice(rv[node])
    rv.pop(node)
    # update remaining values for neihbouring node
    adjList = cg[node]
    for adj in adjList:
        if adj in rv.keys() and color in rv[adj]:
            rv[adj].remove(color)
    #print(adjList)
    return color
```

## Highest Degree Node

```python
# write a method to get the highest degree node
def get_hdnode(cg, solved):
    mnode = ''
    degree = -1
    for k in cg.keys():
        if len(cg[k])  > degree and k not in solved.keys():
            #print(k)
            degree = len(cg[k])
            mnode = k
    return mnode
```

```python
get_hdnode(cgraph, {'SA':'R'})
```
